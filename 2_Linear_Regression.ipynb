# Linear Regression
### Steps in Linear Regression

1. Load the dataset
2. Explore and Visualize the data
3. Split the dataset
4. Fit the Model
5. Evaluate the model
6. Visualise the result

```python
#Import the Necessary Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#step 1: Gererate a synthetic dataset
np.random.seed(42) #For reproducibility
x= 2*np.random.rand(100,1)
y= 7+9*x + np.random.randn(100,1)
```

```python
#step 2 : Visualise the dataset
plt.scatter(x,y,color='blue')
plt.title("Scatter plot of the synthetic dataset")
plt.xlabel("x[Independent variable]")
plt.ylabel("y[Dependent variable]")
plt.show()
```

```python
#step 3 : Split the dataset into training and testing set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#step 4: create and train the Linear Regression Model
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
#step 5 : Make Predicitions using test set
y_pred = model.predict(x_test)
```

```python
#step 6 :  Calculate performance metrics
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_test,y_pred)
```

```python
#Print the coefficients
print(f"Intercept:{model.intercept_[0]:.2f}")
print(f"Slope:{model.coef_[0][0]:.2f}")
```

```python
print(f"Mean Squared Error:{mse:.2f}")
```

```python
print(f"R Squared Error:{r2:.2f}")
```

```python
#Step 7: Visuaize the regresson line
plt.scatter(x_test,y_test,color='blue',label='Actual Values')
plt.plot(x_test,y_pred,color='red',linewidth=2,label='Regression line')
plt.title("Linear Regression Model")
plt.xlabel("x[Independent variable]")
plt.ylabel("y[Dependent variable]")
plt.legend()
plt.show()
```

# Interpretation
1. The Intercept and slope printed show the parameters m and c.
2. A low MSE and High R2 value (close to 1) indicates a good fit
