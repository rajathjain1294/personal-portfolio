```python
#create and save the dataset
import pandas as pd
data = {
    "SquareFootage":[1200,1500,900,1800,2400,1000,2200,1400,1700,2000],
    "Bedrooms":[3,4,2,3,4,2,4,3,3,4],
    "Bathrooms":[2,3,1,2,3,1,2,2,2,3],
    "HousePrice":[250000,310000,180000,400000,520000,200000,490000,275000,380000,450000],
}
```

```python
df=pd.DataFrame(data)
df.to_csv("house_data.csv",index=False)
print("Dataset saved as house_data.csv!")
```

```python
#Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Step 1: Load the dataset
df = pd.read_csv("house_data.csv")
```

```python
#step 2: Data Visualization
#scatter plot
plt.figure(figsize=(12,4))

plt.subplot(1,3,1)

plt.scatter(df["SquareFootage"],df["HousePrice"],alpha=0.7,color="blue")
plt.title("House Price vs Square Footage")
plt.xlabel("Square Footage")
plt.ylabel("House Price")

plt.subplot(1,3,2)

plt.scatter(df["Bedrooms"], df["HousePrice"], alpha=0.7, color="green")
plt.title("House Price vs Bedrooms")
plt.xlabel("Bedrooms")

plt.subplot(1,3,3)
plt.scatter(df["Bathrooms"], df["HousePrice"], alpha=0.7, color="red")
plt.title("House Price vs Bathrooms")
plt.xlabel("Bathrooms")

plt.tight_layout()
plt.show()
```

```python
#correlation heatmap
plt.figure(figsize=(6, 4))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".4f")
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
#Pair Plot

sns.pairplot(df,diag_kind="kde",palette="hus1")
plt.suptitle("Pair plot of features",y=1.02)
plt.show()
```

```python
#Histograms and KDE Plots
plt.figure(figsize=(12,6))

for i, feature in enumerate(["SquareFootage","Bedrooms","Bathrooms","HousePrice"],1):
    plt.subplot(2,2,i)
    sns.histplot(df[feature],kde=True,color=np.random.rand(3,))
    plt.title(f"Distribution of {feature}")

plt.tight_layout()
plt.show()
```
