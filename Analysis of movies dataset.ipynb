# Case Study : Exploring Movie Rating

```python
#Objective : analyse a dataset of movies to gain insights into their genres,ratings, and poplarity
```

#Steps: 1. Load the dataset

        2. Perform basic EDA, including data cleaning and visulization

```python
#Loading the dataset
import pandas as pd
movies=pd.read_csv('movies.csv')
```

```python
#Previewing the data
print(movies.head())
```

```python
#dataset information
print(movies.info())
```

```python
#Missing values
print(movies.isnull().sum())
```

```python
#Top 5 highest rated movies
top_rated = movies.sort_values(by='Rating',ascending=False).head(5)
```

```python
print(top_rated)
```

```python
#genre distribution
genre_counts = movies['Genre'].value_counts()
```

```python
import seaborn as sns
sns.countplot(data=movies, x='Genre', palette='viridis')
```

```python
#Average Rating by genre
avg_rating_by_genre = movies.groupby('Genre')['Rating'].mean()
```

```python
avg_rating_by_genre.plot(kind='bar', color='skyblue')
```

```python
#rating distribution (Histogram)
import matplotlib.pyplot as plt
plt.figure(figsize=(8,5))
sns.histplot(movies['Rating'],kde=True,color='purple',bins=10)
plt.title('Distribution of Movie Rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
#scatter plot:revenue vs rating
plt.figure(figsize=(8, 5))
sns.scatterplot(data=movies, x='Revenue (Millions)', y='Rating', hue='Genre', palette='Set1', s=100, edgecolor='black')
plt.title('Movie Revenue vs Rating')
plt.xlabel('Revenue (Millions)')
plt.ylabel('Rating')
plt.legend(title='Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
#correlation heatmap
corr = movies[['Rating', 'Revenue (Millions)']].corr()
plt.figure(figsize=(7, 5))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Boxplot: Ratings by Genre
plt.figure(figsize=(10, 6))
sns.boxplot(data=movies, x='Genre', y='Rating', palette='Set2')
plt.title('Ratings Distribution by Genre')
plt.xticks(rotation=45)
plt.show()
```

```python
#rating vs year (line plot)
# Rating vs Year (Line Plot)
average_rating_by_year = movies.groupby('Year')['Rating'].mean()
plt.figure(figsize=(10, 6))
average_rating_by_year.plot(kind='line', marker='o', color='green', lw=2)
plt.title('Average Rating vs Year')
plt.xlabel('Year')
plt.ylabel('Average Rating')
plt.grid(True)
plt.show()
```

```python
#Pie chart of genre distribution
plt.figure(figsize=(7, 7))
genre_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='Set3')
plt.title('Genre Distribution')
plt.ylabel('')
plt.show()
```
